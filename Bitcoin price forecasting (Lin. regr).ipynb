```python
#Q1
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error, mean_squared_error
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('Bitcoin.csv')
# Target col
data['Target'] = (data['Low'] + data['High']) / 2
 #preprocess
data['Date'] = pd.to_datetime(data['Date'])
data['Date'] = data['Date'].astype(np.int64)
```

```python
#Drop
data = data.drop(['Name', 'Symbol', 'Open', 'Close'], axis=1)

#Split
X = data.drop(['Target'], axis=1)
y = data['Target']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

#Lin Regr
model = LinearRegression()
model.fit(X_train, y_train)
```

```python
#MAE & MSE
predictions = model.predict(X_test)
mae = mean_absolute_error(y_test, predictions)
mse = mean_squared_error(y_test, predictions)
print("Mean Absolute Error:", mae)
print("Mean Squared Error:", mse)

# Lin Regr Line
plt.scatter(X_test['Date'], y_test, color='black')
plt.plot(X_test['Date'], predictions, color='green', linewidth=2)
plt.title('Bitcoin Price Forecasting / Linear Regression')
plt.xlabel('Date')
plt.ylabel('Bitcoin Price')
plt.show()
```
